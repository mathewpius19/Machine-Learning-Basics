```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nba_csv.csv")
df.head()
#df.describe()
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
df["Salary"]=df["Salary"].fillna(value=df["Salary"].mean())
df.isnull().any()
```

```python
#df["College"] = df["College"].dropna()
#df.isnull().any()
nans = lambda df: df[df.isnull().any(axis=1)]
nans(df)
```

```python
df = df.dropna(axis=0)
df.isnull().any()
```

```python
#use plot function for dataframe to plot graph
#Default line graph is plotted
df.plot(x="Name", y="Age")
```

```python
df[df["Team"] =="Boston Celtics"]
```

```python
df_age = df[df["Team"] =="Boston Celtics"].groupby("Age")["Name"].count().reset_index().rename(columns = {"Name":"Count"})
print(df_age)
df_age.plot.pie(x="Age", y="Count")
```

```python
df_team = df.groupby("Team").sum()#.reset_index()#.sort_values(ascending=False)
print(df_team)
```

```python
df_team.sort_values("Salary").plot.barh(y="Salary")#, figsize=(10,10))
```

```python
df_team=df_team.reset_index()
```

```python
#add column to df_team having count of number of players
df_count = df.groupby("Team")["Name"].count().reset_index().rename(columns={"Name":"Count"})
print(df_count.head(10))
#df_count = df_count.sort_values(by="Count").reset_index()
#print(df_count.head(10))
```

```python
df_team["Count"] = df_count["Count"].sort_values()  #could cause issue since we are joining on index by defult and it could be different
print(df_team)
df_team = df_team.drop("Count", axis=1)
print(df_team)
```

### Merging two dataframes

```python
merged_df = pd.merge(df_team, df_count, how="left",left_on="Team",right_on="Team")
print(merged_df)
```

```python
merged_df.plot.bar(x="Team",y=["Count","Age"], figsize=(10,10))
```

```python
merged_df["Average_age"] = merged_df["Age"]/merged_df["Count"]
```

```python
merged_df.plot.bar(x="Team",y=["Count", "Average_age"], figsize=(10,10))
```
